```python
left = pd.DataFrame({'key':  ['K0','K8', 'K2', 'K3'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', "K2", "K3"],
                     "C": ['C0', 'C1', 'C2', 'C3'],
                     "D": ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right, on='key') #merging on the column 'key'
```

```python
pd.merge(left, right, on='key', how='left')#merging on the column key and taking whole dataset of left 
```

```python
pd.merge(left, right, on='key', how='right')#merging on the column key and taking whole dataset of right 
```

```python
pd.merge(left, right, on='key', how='outer')#merging on the column key and considering whole dataset   as output
```

```python
pd.merge(left, right, on='key', how='inner')
```

```python
df1.join(df2)
```
